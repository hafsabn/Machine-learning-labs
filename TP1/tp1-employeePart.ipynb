```python
import pandas as pd
file = pd.read_csv("employee_salaries.csv")
print(file)
```

```python
# create the plot
import matplotlib.pyplot as plt
plt.xlabel("years of experience")
plt.ylabel("salaries")
plt.title("the salary of employees according to their years of experience")
# plt.grid()
plt.scatter(file['Years_of_Experience'], file['Salary'], color='blue')
```
